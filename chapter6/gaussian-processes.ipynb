# Gaussian Processes
## The prior function $y$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import multivariate_normal, seed
from sklearn.preprocessing import PolynomialFeatures
```

```python
%config InlineBackend.figure_format = "retina"
np.set_printoptions(precision=3, suppress=True)
```

```python
x = np.linspace(-1, 1, 100)
```

## Directly Defining the kernel Matrix

```python
seed(314)
E = np.zeros_like(x)
phi = PolynomialFeatures(degree=6).fit_transform(x[:, np.newaxis])
Phi = phi @ phi.T
plt.plot(np.random.multivariate_normal(E, Phi, size=5).T);
```

## Indirectly Defining the kernel Matrix

### Gaussian Kernel Process

```python
seed(314)
sigma = 0.2
E = np.zeros_like(x)
K = np.exp(-(x[:, np.newaxis] - x[np.newaxis, :])**2 / (2 * sigma))
plt.plot(np.random.multivariate_normal(E, K, size=5).T);
```

### Ornstein-Uhlenbeck Process

```python
seed(314)
theta = 1.2
K = np.exp(-theta * np.abs(x[:, np.newaxis] - x[np.newaxis, :]))
plt.plot(np.random.multivariate_normal(E, K, size=5).T);
```
