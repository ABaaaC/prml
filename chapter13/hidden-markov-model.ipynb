```python
from scipy.stats import multivariate_normal
```

```python
c1 = np.array([[0.8, 0.2]]).T
sigma2 = 0.2
S1 = np.exp(-(c1 - c1.T) ** 2 / 4)
S1
```

```python
N1 = multivariate_normal(mean=[0.2, 0.2], cov=S1)
```

```python
Xgrid = np.mgrid[0:1:0.01, 0:1:0.01]

Z1 = np.apply_along_axis(N1.pdf, 0, Xgrid)
```

```python
plt.contour(*Xgrid, Z1)
plt.xlim(0, 1)
plt.ylim(0, 1)
```
